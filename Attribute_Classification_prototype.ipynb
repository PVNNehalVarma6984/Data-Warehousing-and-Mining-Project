```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Indian_Movies.csv")
df.head()
```

```python
print(df.dtypes)
```

```python
plt.figure(figsize=(14, 8))
plt.bar(df.columns, df.nunique(), color='skyblue')
plt.xlabel('Columns')
plt.ylabel('Number of Unique Values')
plt.title('Unique Values Count in Each Column')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
binary_vars = [col for col in df.columns if df[col].nunique() == 2]
print("Binary Variables:", binary_vars)

sorted_columns = df.nunique().sort_values()

columns_with_fewest_unique_values = sorted_columns.head(2).index.tolist()

if len(columns_with_fewest_unique_values) > 0:
    columns_with_fewest_unique_values = [col for col in columns_with_fewest_unique_values if col not in binary_vars]
    print("Ordinal Values:", columns_with_fewest_unique_values)
else:
    print("No columns with unique values.")

numerical_vars = df.select_dtypes(include='number').columns.tolist()
print("Numerical Values:", numerical_vars)

categorical_vars = df.select_dtypes(include='object').columns.tolist()

categorical_vars = [col for col in categorical_vars if col not in columns_with_fewest_unique_values]
print("Nominal Values:", categorical_vars)

binary_vars = [col for col in df.columns if df[col].nunique() == 2]
print("Binary Variables:", binary_vars)
```
